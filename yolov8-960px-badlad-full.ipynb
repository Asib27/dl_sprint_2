```python
from ultralytics import YOLO

HOURS = 8.6

def train():
    model = YOLO("yolov8m-seg.pt")
    model.train(
        data="badlad.yaml",
        epochs=15,
        imgsz=960,
        batch=4,
        seed=3000,
        overlap_mask=False,
        fliplr=0.0,
        close_mosaic=20,
        show_labels=False,
        conf=0.25,
        iou=0.45,
        max_det=1000,
        val=False,
        half=True,
    )
```

```python
train()
```
